```python
def process_patient_timeseries(df, patient_code, experiment_params, min_duration=10):

    temp_df = df[df["PATIENT_CODE"] == patient_code].reset_index(drop=True)
    subdataframes = split_continuous_timeseries(temp_df, min_duration=min_duration)
    print(subdataframes)
    X, y, coeff_names = [], [], []
    for subdf in subdataframes:
        ppg_values = subdf["PPG_ARRAY"].values
        time_series = preprocess_ppg(ppg_values)

        filtered_series = apply_lowpass_filter(time_series, experiment_params['filter'])

        features, coeff_names = extract_features(filtered_series, experiment_params['features'], experiment_params['bubble'])
        print(coeff_names)
        
        label = 0 if subdf["PATIENT_CODE"].values[0] == 7 else 1
        X.append(features)
        y.append(label)

    return X, y, coeff_names
```

```python
def take_data_for_two(df, unique_patients=[7, 17], experiment_params={}):
    
    
    X, y = [], []
    for patient_code in unique_patients:
        patient_X, patient_y, coeff_names = process_patient_timeseries(df, patient_code, experiment_params)
        X.extend(patient_X)
        y.extend(patient_y)
    
    return np.array(X), np.array(y), coeff_names
```
